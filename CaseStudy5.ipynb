```python
from openpyxl import load_workbook
import matplotlib.pyplot as plt

file_path = '/content/Marks_secA.xlsx'
workbook = load_workbook(filename=file_path)
sheet = workbook.active

names = []
marks = []

for i, row in enumerate(sheet.iter_rows(min_row=2, values_only=True), start=2):
    name = row[1]
    mark = row[6]
    try:
        mark_value = float(mark)
        names.append(name)
        marks.append(mark_value)
    except (TypeError, ValueError):
        print(f"Skipping row {i} due to invalid mark: {mark}")

if not marks:
    print("No valid marks found. Please check Excel column format.")
else:
    print("\nValid marks found. Proceeding with plotting...")
    percentages = [(m / 15) * 100 for m in marks]

    greater_75 = [(names[i], percentages[i]) for i in range(len(percentages)) if percentages[i] > 75]
    between_60_75 = [(names[i], percentages[i]) for i in range(len(percentages)) if 60 <= percentages[i] <= 75]
    less_60 = [(names[i], percentages[i]) for i in range(len(percentages)) if percentages[i] < 60]

    print("\n>75% Students:")
    for n, p in greater_75:
        print(f"{n}: {p:.2f}%")

    print("\n60%-75% Students:")
    for n, p in between_60_75:
        print(f"{n}: {p:.2f}%")

    print("\n<60% Students:")
    for n, p in less_60:
        print(f"{n}: {p:.2f}%")

    plt.figure(figsize=(10, 5))
    plt.hist(percentages, bins=range(0, 101, 5), edgecolor='black', color='blue')
    plt.xlabel("Percentage")
    plt.ylabel("Number of Students")
    plt.title("Histogram Plot")
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.scatter(names, percentages, color='blue', edgecolors='black')
    plt.xlabel("Students")
    plt.ylabel("Percentage")
    plt.title("Scatter Plot")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(True)
    plt.show()
```
